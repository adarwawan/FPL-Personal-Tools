```python
# top_1000_managers = get_top_X_managers(1000)
top_10000_managers = get_top_X_managers(n)
player_info = get_player_info()
teams_dict = get_teams_dict()
```

```python
with open("output.txt", "r") as file:
    # Initialize an empty dictionary
    my_dict = {}

    # Read the file line by line
    for line in file:
        # Strip the newline character and any leading/trailing spaces
        line = line.strip()

        # Split the line into key and value (assuming format "key: value")
        key, value = line.split(": ")
        keyInt = int(key)

        # Convert the value to an integer and add it to the dictionary
        my_dict[keyInt] = int(value)

    ownership_percentage = {player_id: (count / n) * 100 for player_id, count in my_dict.items()}
    sorted_ownership = sorted(ownership_percentage.items(), key=lambda item: item[1], reverse=True)
```

```python
# sorted_ownership = calculate_ownership(top_1000_managers, 2, is_captain=False, is_play_only=False)
df = create_ownership_dataframe(sorted_ownership, player_info, teams_dict, 0)
df
```
